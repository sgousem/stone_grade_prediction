```python
import tensorflow as tf 
from tensorflow import keras
model = keras.models.load_model('cnn_stone_new.h5')
#model = load_model('flowercnn.h5')
print(model.summary())
```

```python
from PIL import Image
import numpy as np
import glob,os
def out_class(result):
    if result[0][0]==1.0:
        image_class = 'mm_10'
    elif result[0][1]==1.0:
        image_class = 'mm_20'
    elif result[0][2]==1.0:
        image_class = 'mm_265'
    elif result[0][3]==1.0:
        image_class = 'sand_plast'
    elif result[0][4]==1.0:
        image_class = 'sand_top'
    else:
        image_class = 'Non predicted'
    return image_class
path = os.getcwd()
print(path)
#os.chdir(os.path.join(path,'flower-test'))
#path1 = os.getcwd()
os.chdir('test')
extension = 'jpg'
all_filenames = [i for i in glob.glob('*.{}'.format(extension))]
for name in all_filenames:
    fname = name.split('.')[0]
    test_image=Image.open(name)
    #img_cup=Image.open('../Data/101_ObjectCategories/Test_images/cup-1.jpg')
    #test_image_cup=Image.open('../Data/101_ObjectCategories/Test_images/crab-cup-1.jpg')
    test_image = test_image.resize((200,200))
    #img_cup = img_cup.resize((200,200))
    #test_image_cup = test_image_cup.resize((200,200))
    test_image=np.expand_dims(test_image,axis=0)
    #img_cup=np.expand_dims(img_cup,axis=0)
    #test_image_cup=np.expand_dims(test_image_cup,axis=0)
    #predict_class = model.predict(test_image)
    pre_class = out_class(model.predict(test_image))
    print('{} Image belongs to: "{}" class'.format(fname,pre_class))  
os.chdir(path)
```
